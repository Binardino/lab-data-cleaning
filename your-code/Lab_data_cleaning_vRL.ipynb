# Data Cleaning 

#### 1. Import pandas library.

```python
import pandas as pd
```

#### 2. Import pymysql and sqlalchemy as you have learnt in the lesson of importing/exporting data. 


```python
from sqlalchemy import create_engine

import pymysql
```

#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/dataset/Stats).

```python
driver   = 'mysql+pymysql:'
user     = 'guest'
password = 'relational'
ip       = 'relational.fit.cvut.cz'
database = 'stats'
```

```python
connection_string = f'{driver}//{user}:{password}@{ip}/{database}'
```

```python
engine = create_engine(connection_string)
```

#### 4. Import the users table.

```python
users = pd.read_sql('SELECT * FROM users', engine)

users.head()
```

#### 5. Rename Id column to userId.

```python
users.rename(columns = {"Id":"userID"}, inplace=True)

users.head()
```

#### 6. Import the posts table. 

```python
# users = pd.read_sql('SELECT * FROM users', engine)

posts = pd.read_sql('SELECT * FROM posts', engine)

posts.head()
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
posts.rename(columns = {"Id":"postID", "OwnerUserId":"userID"}, inplace=True)

posts.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userId, ViewCount, CommentCount

```python
users_renamed = pd.DataFrame(users, columns = ['userID', 'Reputation', 'Views', 'UpVotes', 'DownVotes'])

users_renamed
```

```python
posts_renamed = pd.DataFrame(posts, columns = ['postID', 'Score', 'userID', 'ViewCount', 'CommentCount'])

posts_renamed
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merged_tables = pd.DataFrame.merge(users_renamed ,posts_renamed, left_on = 'userID', right_on = 'postID')

merged_tables
```

#### 9. How many missing values do you have in your merged dataframe? On which columns?

```python
#ViewCount counts NaN

merged_tables.ViewCount.isna().sum()
```

```python
merged_tables.ViewCount.value_counts().sum()
```
